```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress
import pandas as pd
from scipy.optimize import curve_fit
```

```python
#miller

miller = []
for h in range (0,5):
    for k in range (0,5):
        for l in range (0,5):
            miller.append([h,k,l])
#print(miller)

arr_miller = np.asarray(miller)[1:]
print(arr_miller)
```

```python
#Task 2

#BCC
#Ahkl and Ihkl

Ai = 1
lamba = 0.154e-9
a = 0.659e-9

#Structure factor

Rv = np.array([[0,0,0],[1/2,1/2,1/2]])


qv = ((2*np.pi))*arr_miller

i = 1j
Ahkl = np.zeros(len(qv), dtype=complex)
for atom_pos in Rv:
    Ahkl += (Ai*np.exp(i*np.dot(qv,atom_pos)))
    
    
print("Ahkl =",np.abs(Ahkl))

Ihkl = np.abs(Ahkl)**2
print("Ihkl =",np.abs(Ihkl))
```

```python
#Bragg angle - 2dhkl = lamba 
  
D_hkl=np.zeros(len(arr_miller))   
for i in range(len(arr_miller)):
    D_hkl[i]=a/np.sqrt(arr_miller[i][0]**2+arr_miller[i][1]**2+arr_miller[i][2]**2)
  
theta = np.arcsin(lamba/(2*D_hkl)) #radians
#print(theta)

#Finding multiplicity

print("Repeated values are:")
Repeated = []
for i in range(len(theta)):
    Repeated.append(np.count_nonzero(theta == theta[i]))
    
print(Repeated)        

#theta index
print("Index values are:")
Index = []
for i in range(len(theta)):
    Index.append(2**np.count_nonzero(arr_miller[i]))
    
print(Index)

#LP
LP = 1/(theta**(5/2))


#Ireal

Ireal = Ihkl*Repeated*Index*LP

Ireal_n = (Ireal/max(Ireal))*100
#print(Ireal_n)
```

```python
#theta in degrees

d_theta = np.degrees(theta)
print("Theta values:")
print(d_theta)
```

```python
data = {'(h)': arr_miller[:, 0], '(k)': arr_miller[:, 1], '(l)': arr_miller[:, 2], 'Ahkl': np.abs(Ahkl), 'Ihkl': np.abs(Ihkl), 'Multiplicity':Repeated, 'Index': Index, 'LP': LP, 'theta': d_theta, 'Ireal (normalised)': Ireal_n}                                        
df = pd.DataFrame(data=data)
df
```

```python
#Plot

xData = d_theta
yData = Ireal_n

fig = plt.figure(figsize = (12,9))
plt.title('Diffraction model (BCC)')
plt.xlabel('Theta (degrees)')
plt.ylabel('Relative counts (Seconds)')
plt.stem(xData, yData, linefmt = "red", bottom = -10,  label = 'BCC')
plt.ylim(0.0, 120)
plt.legend()
plt.grid()
plt.show() 
```

```python
# -*- coding: utf-8 -*-
"""
Created on Tue Oct 20 21:13:32 2020

"""
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import least_squares, curve_fit



impX,impY = np.loadtxt("./"+filename, delimiter = ',', unpack=True) 

fitRange = np.array([startX,stopX])

initParams = np.array([A,x0,w,c])



impEX = xErr*np.ones(len(impX))
impEY = np.sqrt(impY/Dwell)

boolArray = (impX> fitRange[0]) & (impX< fitRange[1])

xdata = impX[boolArray]
ydata = impY[boolArray]

xerror = impEX[boolArray]
yerror = impEY[boolArray]


fig = plt.figure(figsize = (12, 9))
plt.errorbar(impX,impY, impEY, impEX, marker='.', linestyle='-', label='Your data')
plt.errorbar(xdata, ydata, yerror,xerror, marker='.', linestyle='-', label='Selected data for fitting')
plt.legend(loc = 2)
plt.show()

```

```python
def fitFunc(p, x):
    '''
    Fit function
    '''
    f= p[0]*np.exp(-(1/2)*((x-p[1])/p[2])**2)+p[3]

    return f

def fitFuncDiff(p, x):
    '''
    Differential of fit function
    '''
    df= -p[0]*np.exp(-(1/2)*((x-p[1])/p[2])**2)*((x-p[1])/(p[2]**2))
    return df

def calcChiSq(p, x, y, xerr, yerr):
    '''
    Error function for fit
    '''
    e = (y - fitFunc(p, x))/(np.sqrt(yerr**2 + fitFuncDiff(p, x)**2*xerr**2))
    return e



def fitStdError(jacMatrix):

    # Compute covariance
    jMat2 = np.dot(jacMatrix.T, jacMatrix)
    detJmat2 = np.linalg.det(jMat2)
    
    # Prepare output
    output = np.zeros(jMat2.shape[0])
    if detJmat2 < 1E-32:
        print("Value of determinat detJmat2",detJmat2)
        print("Matrix singular, error calculation failed.")
        return output
    else:
        covar = np.linalg.inv(jMat2)
        for i in range(len(output)):
            output[i] = np.sqrt(covar[i, i])
            
        return output

    
    
    
```

```python
pInit = initParams
lBounds = np.zeros(len(initParams))
uBounds = 10*initParams
nPoints = len(xdata)
nPars = len(initParams)

# Run fit
output = least_squares(calcChiSq, pInit, args = (xdata, ydata, xerror, yerror),
                    bounds = (lBounds, uBounds))



# Get least_squares output, stored in array output.x[]
A = output.x[0]
b = output.x[1]
c = output.x[2]
d = output.x[3]

# Get errors from our fits using fitStdError(), defined above
pErrors = fitStdError(output.jac)
d_A = pErrors[0]
d_b = pErrors[1]
d_c = pErrors[2]
d_d = pErrors[3]



# Calculate fitted y-values using our fit parameters and the original fit function
xPlot = np.linspace(np.min(xdata), np.max(xdata), 300)
fitData = fitFunc(output.x, xPlot)

# Make the plot of the data and the fit
fig = plt.figure(figsize = (12, 9))
plt.title('Fit of '+ filename)
plt.xlabel('x')
plt.ylabel('y')
#plt.grid(color = 'g')
plt.errorbar(xdata, ydata, yerror, xerror, marker='.', linestyle='', label='data')
plt.plot(xPlot, fitData, color = 'r', label= 'fit')
plt.show()


# Output fit parameters
print("Fitted parameters: A= {0:.2f}, x0= {1:.2f}, w= {2:.2f}, c= {3:.2f}".format(A, b, c, d))
print("Parameter errors: A= {0:.2f}, x0= {1:.2f}, w= {2:.2f}, c= {3:.2f}".format(d_A, d_b, d_c, d_d))



# Calculate chis**2 per point, summed chi**2 and chi**2/NDF
chiarr = calcChiSq(output.x, xdata, ydata, xerror, yerror)**2
chisq = np.sum(chiarr)
NDF = nPoints - nPars
chisqndf = chisq/NDF
print("ChiSq = {:5.2e}, ChiSq/NDF = {:5.2f}.".format(chisq, chisqndf))


```

```python
#Mean crystallite size (peak 1)
sig = np.radians(0.32) #Width of Gaussian

FWHM = 2*np.sqrt(2*np.log(2))*sig #Full width half maximum

#Bragg angle

B_theta = np.radians(15.79)

L = ((0.9*lamba)/(2*FWHM*np.cos(B_theta)))*1e10

#Errors

sigerr = np.radians(0.20)
b_err = np.radians(0.10)

#FWHM error

FWHMerr = ((0.9*lamba/2*np.cos(B_theta))*(-FWHM**-2)*sigerr)*1e10

#thetaerr

theta_err = ((0.9*lamba/2*FWHM)*(np.sin(B_theta)/np.cos(B_theta)**2)*b_err)*1e10

#Combining errors

L_err = np.sqrt(FWHMerr**2 + theta_err**2)

print("Mean crystallite size =",L,"+-",L_err,"A")
```

```python
pInit = initParams
lBounds = np.zeros(len(initParams))
uBounds = 10*initParams
nPoints = len(xdata)
nPars = len(initParams)

# Run fit
output = least_squares(calcChiSq, pInit, args = (xdata, ydata, xerror, yerror),
                    bounds = (lBounds, uBounds))



# Get least_squares output, stored in array output.x[]
A = output.x[0]
b = output.x[1]
c = output.x[2]
d = output.x[3]

# Get errors from our fits using fitStdError(), defined above
pErrors = fitStdError(output.jac)
d_A = pErrors[0]
d_b = pErrors[1]
d_c = pErrors[2]
d_d = pErrors[3]



# Calculate fitted y-values using our fit parameters and the original fit function
xPlot = np.linspace(np.min(xdata), np.max(xdata), 300)
fitData = fitFunc(output.x, xPlot)

# Make the plot of the data and the fit
fig = plt.figure(figsize = (12, 9))
plt.title('Fit of '+ filename)
plt.xlabel('x')
plt.ylabel('y')
#plt.grid(color = 'g')
plt.errorbar(xdata, ydata, yerror, xerror, marker='.', linestyle='', label='data')
plt.plot(xPlot, fitData, color = 'r', label= 'fit')
plt.show()


# Output fit parameters
print("Fitted parameters: A= {0:.2f}, x0= {1:.2f}, w= {2:.2f}, c= {3:.2f}".format(A, b, c, d))
print("Parameter errors: A= {0:.2f}, x0= {1:.2f}, w= {2:.2f}, c= {3:.2f}".format(d_A, d_b, d_c, d_d))



# Calculate chis**2 per point, summed chi**2 and chi**2/NDF
chiarr = calcChiSq(output.x, xdata, ydata, xerror, yerror)**2
chisq = np.sum(chiarr)
NDF = nPoints - nPars
chisqndf = chisq/NDF
print("ChiSq = {:5.2e}, ChiSq/NDF = {:5.2f}.".format(chisq, chisqndf))
```

```python
# -*- coding: utf-8 -*-
"""
Created on Tue Oct 20 21:13:32 2020

"""
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import least_squares, curve_fit



impX,impY = np.loadtxt("./"+filename, delimiter = ',', unpack=True) 

fitRange = np.array([startX,stopX])

initParams = np.array([A,x0,w,c])



impEX = xErr*np.ones(len(impX))
impEY = np.sqrt(impY/Dwell)

boolArray = (impX> fitRange[0]) & (impX< fitRange[1])

xdata = impX[boolArray]
ydata = impY[boolArray]

xerror = impEX[boolArray]
yerror = impEY[boolArray]


fig = plt.figure(figsize = (12, 9))
plt.errorbar(impX,impY, impEY, impEX, marker='.', linestyle='-', label='Your data')
plt.errorbar(xdata, ydata, yerror,xerror, marker='.', linestyle='-', label='Selected data for fitting')
plt.legend(loc = 2)
plt.show()

```

```python
def fitFunc(p, x):
    '''
    Fit function
    '''
    f= p[0]*np.exp(-(1/2)*((x-p[1])/p[2])**2)+p[3]

    return f

def fitFuncDiff(p, x):
    '''
    Differential of fit function
    '''
    df= -p[0]*np.exp(-(1/2)*((x-p[1])/p[2])**2)*((x-p[1])/(p[2]**2))
    return df

def calcChiSq(p, x, y, xerr, yerr):
    '''
    Error function for fit
    '''
    e = (y - fitFunc(p, x))/(np.sqrt(yerr**2 + fitFuncDiff(p, x)**2*xerr**2))
    return e



def fitStdError(jacMatrix):

    # Compute covariance
    jMat2 = np.dot(jacMatrix.T, jacMatrix)
    detJmat2 = np.linalg.det(jMat2)
    
    # Prepare output
    output = np.zeros(jMat2.shape[0])
    if detJmat2 < 1E-32:
        print("Value of determinat detJmat2",detJmat2)
        print("Matrix singular, error calculation failed.")
        return output
    else:
        covar = np.linalg.inv(jMat2)
        for i in range(len(output)):
            output[i] = np.sqrt(covar[i, i])
            
        return output

    
    
    
```

```python
pInit = initParams
lBounds = np.zeros(len(initParams))
uBounds = 10*initParams
nPoints = len(xdata)
nPars = len(initParams)

# Run fit
output = least_squares(calcChiSq, pInit, args = (xdata, ydata, xerror, yerror),
                    bounds = (lBounds, uBounds))



# Get least_squares output, stored in array output.x[]
A = output.x[0]
b = output.x[1]
c = output.x[2]
d = output.x[3]

# Get errors from our fits using fitStdError(), defined above
pErrors = fitStdError(output.jac)
d_A = pErrors[0]
d_b = pErrors[1]
d_c = pErrors[2]
d_d = pErrors[3]



# Calculate fitted y-values using our fit parameters and the original fit function
xPlot = np.linspace(np.min(xdata), np.max(xdata), 300)
fitData = fitFunc(output.x, xPlot)

# Make the plot of the data and the fit
fig = plt.figure(figsize = (12, 9))
plt.title('Fit of '+ filename)
plt.xlabel('x')
plt.ylabel('y')
#plt.grid(color = 'g')
plt.errorbar(xdata, ydata, yerror, xerror, marker='.', linestyle='', label='data')
plt.plot(xPlot, fitData, color = 'r', label= 'fit')
plt.show()


# Output fit parameters
print("Fitted parameters: A= {0:.2f}, x0= {1:.2f}, w= {2:.2f}, c= {3:.2f}".format(A, b, c, d))
print("Parameter errors: A= {0:.2f}, x0= {1:.2f}, w= {2:.2f}, c= {3:.2f}".format(d_A, d_b, d_c, d_d))



# Calculate chis**2 per point, summed chi**2 and chi**2/NDF
chiarr = calcChiSq(output.x, xdata, ydata, xerror, yerror)**2
chisq = np.sum(chiarr)
NDF = nPoints - nPars
chisqndf = chisq/NDF
print("ChiSq = {:5.2e}, ChiSq/NDF = {:5.2f}.".format(chisq, chisqndf))
```

```python
def fitFunc(x, *p):
    '''
    Fit function
    '''
    f = p[0] + p[1]*x
    return f

def fitFuncDiff(x, *p):
    '''
    Differential of fit function
    '''
    df = p[1]
    return df

def calcChiSq(p, x, y, yerr):
    '''
    Error function for fit
    '''
    e = (y - fitFunc(x, *p))/yerr
    return e
```

```python
# Set initial values of fit parameters
pInit =   [1., 1.]
nPoints = len(xdata)
nPars = 2

# Run fit
popt, pcov = curve_fit(fitFunc, xdata, ydata, p0=pInit)

# Calculate the fit errors on the parameters by the square root of the diagonals of 
# the covariance matrix (don't worry too much about this)
pError = np.sqrt(np.diag(pcov))

# Output fit parameters
print("Fitted parameters m: {0:.7f}, x: {1:.7f}".format(popt[1], popt[0]))
print("Parameter errors: m: {0:.7}, x: {1:.7f}".format(pError[1], pError[0]))


# Calculate chis**2 per point, summed chi**2 and chi**2/NDF
chiarr = calcChiSq(popt, xdata, ydata, yerror)**2
chisq = np.sum(chiarr)
NDF = nPoints - nPars
chisqndf = chisq/NDF
print("ChiSq = {:5.2e}, ChiSq/NDF = {:5.7f}.".format(chisq, chisqndf))

# Calculate fitted y-values using our fit parameters and the original fit function
xPlot = np.linspace(0.9*np.min(xdata), 1.1*np.max(xdata), 100)
fitData = fitFunc(xPlot, *popt)

# Make the plot of the data and the fit
fig = plt.figure(figsize = (8, 6))
plt.title('Sin^2 theta vs N for KCl')
plt.xlabel('N')
plt.ylabel('Sin^2 theta (degrees)')
plt.grid(color = 'g')
plt.errorbar(xdata, ydata, yerror, marker='.', linestyle='', label='data')
plt.plot(xPlot, fitData, color = 'r')
plt.show()
```
